```python
import numpy as np
from stable_baselines3.common.callbacks import BaseCallback

class RewardLogger(BaseCallback):

    def __init__(self, verbose: int = 0):
        super().__init__(verbose)
        self.episode_rewards = []
        self.episode_timesteps = []
        self._current_rewards = None

    def _on_training_start(self) -> None:
        n_envs = self.training_env.num_envs
        self._current_rewards = np.zeros(n_envs, dtype=np.float32)

    def _on_step(self) -> bool:
        rewards = self.locals["rewards"]
        dones   = self.locals["dones"]

        self._current_rewards += rewards

        for i, done in enumerate(dones):
            if done:
                ep_rew = float(self._current_rewards[i])
                self.episode_rewards.append(ep_rew)
                self.episode_timesteps.append(self.num_timesteps)
                self._current_rewards[i] = 0.0

        return True
```

```python
import gymnasium as gym
from stable_baselines3 import SAC

# Make HalfCheetah environment (MuJoCo)
env = gym.make("HalfCheetah-v4")

# Create SAC model
model = SAC(
    "MlpPolicy",
    env,
    verbose=0,
    learning_rate=3e-4,
    buffer_size=1_000_000,
    batch_size=256,
    tau=0.005,
    gamma=0.99,
    device="cuda",
    train_freq=1,
    gradient_steps=1,
    learning_starts=10_000,
    ent_coef="auto",
)

# Create RewardLogger callback
reward_logger = RewardLogger(verbose=0)

# Train the agent
TOTAL_TIMESTEPS = 1_000_000
model.learn(
    total_timesteps=TOTAL_TIMESTEPS,
    progress_bar=True,
    callback=reward_logger,
)

model.save("sac_halfcheetah_auto_1000k")
env.close()
```

```python
import gymnasium as gym
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import imageio_ffmpeg
import matplotlib as mpl


mpl.rcParams["animation.ffmpeg_path"] = imageio_ffmpeg.get_ffmpeg_exe()

# Re-create env in rgb_array mode for video
eval_env = gym.make("HalfCheetah-v4", render_mode="rgb_array")

N_STEPS = 500   # how many steps to record
frames = []

obs, _ = eval_env.reset()
for _ in range(N_STEPS):
    # Use the trained SAC model
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = eval_env.step(action)

    # Grab a frame
    frame = eval_env.render()
    frames.append(frame)

    if terminated or truncated:
        obs, _ = eval_env.reset()

eval_env.close()

# Build matplotlib animation
fig, ax = plt.subplots()
img = ax.imshow(frames[0])
ax.axis("off")

def animate(i):
    img.set_data(frames[i])
    return [img]

ani = animation.FuncAnimation(fig, animate, frames=len(frames), interval=30)


plt.close(fig)  # so Jupyter doesn't show the static image

# Show inline HTML5 video
HTML(ani.to_html5_video())
```

```python
import matplotlib.pyplot as plt
import numpy as np

rewards = np.array(reward_logger.episode_rewards)
timesteps = np.array(reward_logger.episode_timesteps)

plt.figure(figsize=(8, 5))
plt.plot(timesteps, rewards, label="SAC (raw)")

# Smooth a bit
window = 5
if len(rewards) >= window:
    smooth = np.convolve(rewards, np.ones(window)/window, mode="valid")
    smooth_steps = timesteps[window-1:]
    plt.plot(smooth_steps, smooth, label=f"MA({window})", linewidth=2)

plt.xlabel("Timesteps")
plt.ylabel("Episode reward")
plt.title("HalfCheetah-v4: SAC learning curve")
plt.legend()
plt.grid(True)
plt.show()
```

SAC vs PPO
In my HalfCheetah-v4 experiments, SAC did better than PPO. It reached higher average rewards and got there faster in terms of timesteps. PPO still improved, but its curve was noisier and seemed to plateau earlier than SAC.

Entropy temperature effect
When I used a higher entropy temperature (larger ent_coef), the agent explored more but learned more slowly and the rewards were less stable. With lower entropy, it improved faster at the beginning but sometimes got stuck in a worse local solution. Using ent_coef="auto" gave the most balanced behavior in terms of exploration and stable convergence.

Exploration and motion behavior
At higher entropy, the HalfCheetah’s movement looked more random at first—lots of weird leg motions and sometimes even going backward. As entropy went down or the training progressed, the motion became smoother and more consistent, like a proper running gait. Compared to PPO, the SAC policy looked more stable and less jerky once it was trained.
